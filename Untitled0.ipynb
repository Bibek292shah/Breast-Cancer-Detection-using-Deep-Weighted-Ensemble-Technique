```python
# ===== CELL: Robust AMP monkey-patch + safe scaler creation =====
import inspect, torch, warnings, types

print("Patching AMP API to be robust across torch versions...")

# --- Safe GradScaler wrapper factory ---
def _make_safe_grad_scaler():
    # prefer torch.amp.GradScaler if available
    GradScaler = getattr(torch, "amp", None) and getattr(torch.amp, "GradScaler", None)
    if GradScaler is None:
        # fallback to torch.cuda.amp.GradScaler if present
        GradScaler = getattr(torch.cuda.amp, "GradScaler", None)
    if GradScaler is None:
        raise RuntimeError("No GradScaler available in this torch build.")

    sig = inspect.signature(GradScaler.__init__)
    params = list(sig.parameters.keys())

    def safe_grad_scaler(*args, **kwargs):
        # Try passing device_type if it's supported
        try:
            if "device_type" in params:
                # pass device_type if not already provided
                if "device_type" not in kwargs:
                    kwargs2 = dict(kwargs)
                    kwargs2["device_type"] = ('cuda' if torch.cuda.is_available() else 'cpu')
                else:
                    kwargs2 = kwargs
                return GradScaler(*args, **kwargs2)
            else:
                # constructor doesn't accept device_type -> call without it
                return GradScaler(*args, **kwargs)
        except TypeError:
            # Last-resort: call without kwargs
            return GradScaler(*args)

    # Wrap to look like a class (callable returning an instance)
    return safe_grad_scaler

# --- Safe autocast wrapper ---
def _make_safe_autocast():
    # prefer torch.amp.autocast, fallback to torch.cuda.amp.autocast
    autocast_fn = getattr(torch, "amp", None) and getattr(torch.amp, "autocast", None)
    if autocast_fn is None:
        autocast_fn = getattr(torch.cuda.amp, "autocast", None)
    if autocast_fn is None:
        raise RuntimeError("No autocast available in this torch build.")

    # Create a context manager wrapper that supports autocast(enabled=...) and device_type selection
    class _AutocastWrapper:
        def __init__(self, **outer_kwargs):
            # store outer kwargs (not used much)
            self.outer_kwargs = outer_kwargs

        def __call__(self, *args, **kwargs):
            # handle legacy 'enabled' kw
            enabled = kwargs.pop("enabled", None)
            if enabled is False:
                # return a no-op context manager
                class _Noop:
                    def __enter__(self): return None
                    def __exit__(self, exc_type, exc, tb): return False
                return _Noop()
            # Determine device_type if new API expects it
            try:
                sig = inspect.signature(autocast_fn)
                if "device_type" in sig.parameters:
                    device_type = 'cuda' if torch.cuda.is_available() else 'cpu'
                    return autocast_fn(device_type=device_type, *args, **kwargs)
            except (ValueError, TypeError):
                pass
            # fallback: call with args/kwargs
            return autocast_fn(*args, **kwargs)

    return _AutocastWrapper()

# --- Apply safe wrappers to torch.cuda.amp namespace (so older code using torch.cuda.amp.* keeps working) ---
try:
    safe_grad = _make_safe_grad_scaler()
    safe_autocast = _make_safe_autocast()

    # expose in torch.cuda.amp for backward compatibility
    torch.cuda.amp.GradScaler = safe_grad
    torch.cuda.amp.autocast = safe_autocast

    # and also expose in torch.amp for newer code
    if hasattr(torch, "amp"):
        torch.amp.GradScaler = getattr(torch, "amp").GradScaler if getattr(torch, "amp", None) and hasattr(torch.amp, "GradScaler") else safe_grad
        # for autocast, keep torch.amp.autocast as-is (it's the real one)
except Exception as e:
    warnings.warn(f"Could not apply AMP patch cleanly: {e}")

# --- Create a usable scaler instance (safe) ---
try:
    # Prefer torch.amp.GradScaler if present and callable
    if hasattr(torch, "amp") and hasattr(torch.amp, "GradScaler"):
        try:
            scaler = torch.amp.GradScaler()
        except TypeError:
            # try without args
            scaler = torch.amp.GradScaler()
    else:
        # fallback to our safe wrapper which returns an instance
        scaler = torch.cuda.amp.GradScaler()
except Exception as e:
    # As a last resort, set scaler to None and let training code handle absence
    scaler = None
    warnings.warn(f"Could not create GradScaler instance: {e}. Continuing with scaler=None (no AMP).")

print("Scaler created:", type(scaler), "CUDA available:", torch.cuda.is_available())
```

```python
# ============================================
# CELL 6 — SWIN V2 MODEL + TRAINING (CLEAN)
# ============================================
import timm
import torch.nn as nn
import torch.optim as optim
from tqdm.notebook import tqdm
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import time, numpy as np, os

# ---- try to load H-type SwinV2, fallback to large ----
MODEL_NAME = None
for name in ["swinv2_huge_window12to16_192to256_22kto1k_ft",
             "swinv2_huge_window12_192_22k",
             "swinv2_large_window12_192_22k"]:
    try:
        model = timm.create_model(name, pretrained=True, num_classes=len(train_dataset.classes))
        MODEL_NAME = name
        print("Loaded:", name)
        break
    except:
        print("Could not load:", name)

if MODEL_NAME is None:
    raise RuntimeError("SwinV2 model not found in timm!")

# enable checkpointing for huge models
try:
    model.set_grad_checkpointing(True)
except:
    pass

model.to(device)

# ---- get correct IMAGE SIZE from model ----
cfg = model.default_cfg
IMG_SIZE = int(cfg["input_size"][1])
print("Model expects:", IMG_SIZE)

# ---- replace transforms with correct size ----
mean = [0.485,0.456,0.406]
std  = [0.229,0.224,0.225]

train_dataset.transform = transforms.Compose([
    transforms.RandomResizedCrop(IMG_SIZE, scale=(0.8,1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(15),
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
])

test_dataset.transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
])

train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True, num_workers=2)
test_loader  = DataLoader(test_dataset,  batch_size=4, shuffle=False, num_workers=2)

# ---- compute class weights ----
from collections import Counter
cnt = Counter(train_dataset.targets)
class_counts = np.array([cnt[i] for i in range(len(train_dataset.classes))], dtype=np.float32)
weights = class_counts.sum() / (len(class_counts) * class_counts)
weights = torch.tensor(weights, dtype=torch.float32).to(device)
print("Class counts:", class_counts, "Weights:", weights.cpu().numpy())

criterion = nn.CrossEntropyLoss(weight=weights)

optimizer = optim.AdamW(model.parameters(), lr=2e-4, weight_decay=1e-2)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=25)
scaler = torch.cuda.amp.GradScaler()

EPOCHS = 25
ACCUM = 4
best_acc = 0
stale = 0
PATIENCE = 5

os.makedirs("checkpoints", exist_ok=True)

def train_one_epoch(epoch):
    model.train()
    total_loss, preds_all, targs_all = 0, [], []
    optimizer.zero_grad()
    pbar = tqdm(enumerate(train_loader), total=len(train_loader), desc=f"E{epoch+1}")

    for i, (x, y) in pbar:
        x, y = x.to(device), y.to(device)
        with torch.cuda.amp.autocast():
            yhat = model(x)
            loss = criterion(yhat, y) / ACCUM

        scaler.scale(loss).backward()

        if (i+1) % ACCUM == 0:
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()

        total_loss += loss.item() * ACCUM
        preds_all += yhat.argmax(1).cpu().tolist()
        targs_all += y.cpu().tolist()

        pbar.set_postfix(loss=total_loss/(i+1),
                         acc=accuracy_score(targs_all, preds_all))

    return total_loss/len(train_loader), accuracy_score(targs_all, preds_all)


def eval_model():
    model.eval()
    total_loss, preds_all, targs_all = 0, [], []

    with torch.no_grad():
        for x, y in tqdm(test_loader, total=len(test_loader), desc="Eval"):
            x, y = x.to(device), y.to(device)
            with torch.cuda.amp.autocast():
                yhat = model(x)
                loss = criterion(yhat, y)

            total_loss += loss.item()
            preds_all += yhat.argmax(1).cpu().tolist()
            targs_all += y.cpu().tolist()

    return total_loss/len(test_loader), accuracy_score(targs_all, preds_all), preds_all, targs_all


# ---- TRAINING LOOP ----
for epoch in range(EPOCHS):
    start = time.time()
    tr_loss, tr_acc = train_one_epoch(epoch)
    val_loss, val_acc, pv, tv = eval_model()
    scheduler.step()

    print(f"[E{epoch+1}] TrainAcc={tr_acc:.4f}  ValAcc={val_acc:.4f}")

    if val_acc > best_acc:
        best_acc = val_acc
        stale = 0
        path = f"checkpoints/best_{MODEL_NAME}_acc{val_acc:.4f}.pth"
        torch.save(model.state_dict(), path)
        print("Saved:", path)
    else:
        stale += 1

    if stale >= PATIENCE:
        print("Early stopping triggered.")
        break

print("BEST VAL ACC:", best_acc)
print(classification_report(tv, pv, target_names=train_dataset.classes))
print("Confusion:\n", confusion_matrix(tv, pv))

torch.save(model.state_dict(), "final_swinv2_busi.pth")
```

```python
# Cell 7 — dataloader makers, model builders, and training loop
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms, models, datasets
from torch.utils.data import DataLoader
from sklearn.metrics import accuracy_score, classification_report
from tqdm.notebook import tqdm

torch.backends.cudnn.benchmark = True  # speedup for fixed-size inputs where applicable

# Reuse train_idx/test_idx produced in your cell 5
# train_idx, test_idx, DATA_ROOT, BATCH_SIZE, IMG_SIZE are expected to exist.

# Make model-specific data loaders
def make_dataloaders_for_model(model_name, batch_size=BATCH_SIZE, alex_img_size=224):
    """
    Returns train_loader, test_loader, num_classes
    - For AlexNet uses 224x224 resizing (pretrained AlexNet expects standard sizes).
    - For ResNet18 and MobileNetV2 we use IMG_SIZE from your cell 5 (could be 600).
    """
    # base val transform (for resnet / mobilenet)
    val_transform_general = transforms.Compose([
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
    ])
    train_transform_general = transforms.Compose([
        transforms.Resize((IMG_SIZE, IMG_SIZE)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(15),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
    ])

    # alex transforms (224)
    alex_val_transform = transforms.Compose([
        transforms.Resize((alex_img_size, alex_img_size)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
    ])
    alex_train_transform = transforms.Compose([
        transforms.Resize((alex_img_size, alex_img_size)),
        transforms.RandomChoice([
            transforms.RandomHorizontalFlip(p=1.0),
            transforms.RandomVerticalFlip(p=1.0),
            transforms.RandomRotation((90,90)),
            transforms.RandomAffine(degrees=0)
        ]),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
    ])

    if model_name.lower().startswith("alex"):
        train_ds = datasets.ImageFolder(root=DATA_ROOT, transform=alex_train_transform)
        val_ds = datasets.ImageFolder(root=DATA_ROOT, transform=alex_val_transform)
    else:
        train_ds = datasets.ImageFolder(root=DATA_ROOT, transform=train_transform_general)
        val_ds = datasets.ImageFolder(root=DATA_ROOT, transform=val_transform_general)

    # Restrict samples to the train/test indices (keep same split as you created)
    # train_idx and test_idx come from your previous split
    train_ds.samples = [train_ds.samples[i] for i in train_idx]
    train_ds.targets = [train_ds.targets[i] for i in train_idx]
    val_ds.samples = [val_ds.samples[i] for i in test_idx]
    val_ds.targets = [val_ds.targets[i] for i in test_idx]

    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)
    val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False, num_workers=2, pin_memory=True)

    num_classes = len(train_ds.classes)
    return train_loader, val_loader, num_classes

# Model builder (pretrained, adapt final head)
def build_model(model_name, num_classes, pretrained=True):
    model_name = model_name.lower()
    if model_name == "alexnet":
        model = models.alexnet(pretrained=pretrained)
        # Replace classifier to match num_classes
        in_features = model.classifier[6].in_features
        model.classifier[6] = nn.Linear(in_features, num_classes)
    elif model_name == "resnet18":
        model = models.resnet18(pretrained=pretrained)
        in_features = model.fc.in_features
        model.fc = nn.Linear(in_features, num_classes)
    elif model_name == "mobilenetv2":
        model = models.mobilenet_v2(pretrained=pretrained)
        in_features = model.classifier[1].in_features
        model.classifier[1] = nn.Linear(in_features, num_classes)
    else:
        raise ValueError("Unsupported model_name")
    return model

# Training + evaluation loop with mixed precision and checkpoint saving
def train_and_eval(model_name, epochs=10, lr=2e-4, weight_decay=1e-4, print_report=True):
    print(f"\n=== Training {model_name} for {epochs} epochs ===")
    train_loader, val_loader, num_classes = make_dataloaders_for_model(model_name)
    model = build_model(model_name, num_classes).to(device)

    # optimizer + scheduler
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs, eta_min=1e-6)

    criterion = nn.CrossEntropyLoss()

    scaler = torch.cuda.amp.GradScaler(enabled=(device.type == "cuda"))

    best_val_acc = 0.0
    best_local_path = None
    history = {"train_loss": [], "val_loss": [], "val_acc": []}

    for epoch in range(1, epochs + 1):
        model.train()
        running_loss = 0.0
        train_steps = 0
        pbar = tqdm(train_loader, desc=f"{model_name} Epoch {epoch}/{epochs}", leave=False)
        for images, labels in pbar:
            images = images.to(device, non_blocking=True)
            labels = labels.to(device, non_blocking=True)

            optimizer.zero_grad()
            with torch.cuda.amp.autocast(enabled=(device.type == "cuda")):
                outputs = model(images)
                loss = criterion(outputs, labels)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            running_loss += loss.item()
            train_steps += 1
            pbar.set_postfix({"loss": running_loss / train_steps})

        train_loss = running_loss / max(1, train_steps)
        scheduler.step()

        # Validation
        model.eval()
        all_preds = []
        all_labels = []
        val_loss_sum = 0.0
        val_steps = 0
        with torch.no_grad():
            for images, labels in val_loader:
                images = images.to(device, non_blocking=True)
                labels = labels.to(device, non_blocking=True)
                with torch.cuda.amp.autocast(enabled=(device.type == "cuda")):
                    outputs = model(images)
                    loss = criterion(outputs, labels)
                val_loss_sum += loss.item()
                val_steps += 1
                preds = torch.argmax(outputs, dim=1).detach().cpu().numpy().tolist()
                labs = labels.detach().cpu().numpy().tolist()
                all_preds.extend(preds)
                all_labels.extend(labs)

        val_loss = val_loss_sum / max(1, val_steps)
        val_acc = accuracy_score(all_labels, all_preds)
        history["train_loss"].append(train_loss)
        history["val_loss"].append(val_loss)
        history["val_acc"].append(val_acc)

        print(f"{model_name} Epoch {epoch}: train_loss={train_loss:.4f} val_loss={val_loss:.4f} val_acc={val_acc:.4f}")

        # Save checkpoint locally every epoch
        state = {
            "epoch": epoch,
            "model_state_dict": model.state_dict(),
            "optimizer_state_dict": optimizer.state_dict(),
            "val_acc": val_acc,
            "val_loss": val_loss,
            "history": history,
            "model_name": model_name
        }
        local_ckpt = save_epoch_checkpoint(state, model_name, epoch)

        # If best, copy to drive (best snapshot)
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_local_path = local_ckpt
            copy_best_to_drive(best_local_path, model_name)

    # Final evaluation classification report on test set
    if print_report:
        print(f"\nFinal evaluation for {model_name} (best val acc={best_val_acc:.4f}):")
        print(classification_report(all_labels, all_preds, target_names=train_loader.dataset.classes, zero_division=0))

    return {
        "history": history,
        "best_val_acc": best_val_acc,
        "best_local_path": best_local_path
    }
```

```python
# Cell 7 — build EfficientNet-B7, optimizer, scheduler, AMP scaler
import timm

NUM_CLASSES = len(dataset_all.classes)
print("Classes:", dataset_all.classes, "NUM_CLASSES:", NUM_CLASSES)

# Create model
model = timm.create_model('tf_efficientnet_b7', pretrained=True, num_classes=NUM_CLASSES)
# If you're memory-constrained, you can replace with smaller model such as 'tf_efficientnet_b4' or 'tf_efficientnet_b3'

# Move to device and wrap if multi-gpu
model = model.to(device)
if torch.cuda.device_count() > 1:
    model = nn.DataParallel(model)
    print("Wrapped model in DataParallel")

# Loss, optimizer, scheduler
criterion = nn.CrossEntropyLoss()
LEARNING_RATE = 1e-4
WEIGHT_DECAY = 1e-4
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)

EPOCHS = 18          # you can increase (monitor validation) — lower if OOM
steps_per_epoch = max(1, len(train_loader))
MAX_LR = 2e-4
scheduler = optim.lr_scheduler.OneCycleLR(
    optimizer, max_lr=MAX_LR, steps_per_epoch=steps_per_epoch, epochs=EPOCHS,
    pct_start=0.1, div_factor=10.0, final_div_factor=100.0
)

scaler = torch.cuda.amp.GradScaler()

print("Model created. Params:", sum(p.numel() for p in model.parameters())/1e6, "M")
```

```python
# Cell 8 — helper functions (save/load), reproducible seeds
import random
# ensure SEED exists (should from earlier fix cell)
try:
    SEED
except NameError:
    SEED = 42

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

def save_checkpoint(model, optimizer, epoch, path):
    state = {
        "epoch": epoch,
        "model_state": (model.module.state_dict() if hasattr(model, "module") else model.state_dict()),
        "optimizer_state": optimizer.state_dict()
    }
    torch.save(state, path)

def load_checkpoint(path, model, map_location=None):
    ck = torch.load(path, map_location=map_location)
    model_state = ck["model_state"]
    # load into model (handles DataParallel)
    if hasattr(model, "module"):
        model.module.load_state_dict(model_state)
    else:
        model.load_state_dict(model_state)
    return ck.get("epoch", None)
```

```python
# Cell 9 — training loop with AMP, progress bars, best-model checkpointing
BEST_PATH = "best_efficientnet_b7.pth"
best_val_acc = 0.0
history = {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": []}

for epoch in range(1, EPOCHS + 1):
    model.train()
    running_loss = 0.0
    running_corrects = 0
    total_examples = 0

    loop = tqdm(train_loader, desc=f"Epoch {epoch}/{EPOCHS} Train", leave=False)
    for images, targets in loop:
        images = images.to(device, non_blocking=True)
        targets = targets.to(device, non_blocking=True)

        optimizer.zero_grad()
        with torch.cuda.amp.autocast():
            outputs = model(images)
            loss = criterion(outputs, targets)

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=5.0)  # helps stability
        scaler.step(optimizer)
        scaler.update()

        # OneCycleLR step per batch
        try:
            scheduler.step()
        except Exception:
            pass

        preds = torch.argmax(outputs.detach(), dim=1)
        batch_size = images.size(0)
        running_loss += loss.item() * batch_size
        running_corrects += torch.sum(preds == targets).item()
        total_examples += batch_size
        loop.set_postfix(loss=(running_loss/total_examples), acc=(running_corrects/total_examples))

    epoch_train_loss = running_loss / total_examples
    epoch_train_acc = running_corrects / total_examples

    # Validation
    model.eval()
    val_loss = 0.0
    val_corrects = 0
    val_total = 0
    all_targets = []
    all_probs = []

    with torch.no_grad():
        for images, targets in tqdm(test_loader, desc=f"Epoch {epoch}/{EPOCHS} Val", leave=False):
            images = images.to(device, non_blocking=True)
            targets = targets.to(device, non_blocking=True)

            with torch.cuda.amp.autocast():
                outputs = model(images)
                loss = criterion(outputs, targets)
                probs = F.softmax(outputs, dim=1)

            batch_size = images.size(0)
            val_loss += loss.item() * batch_size
            preds = torch.argmax(probs, dim=1)
            val_corrects += torch.sum(preds == targets).item()
            val_total += batch_size

            all_targets.extend(targets.cpu().numpy().tolist())
            all_probs.extend(probs.cpu().numpy().tolist())

    epoch_val_loss = val_loss / val_total
    epoch_val_acc = val_corrects / val_total

    history["train_loss"].append(epoch_train_loss)
    history["val_loss"].append(epoch_val_loss)
    history["train_acc"].append(epoch_train_acc)
    history["val_acc"].append(epoch_val_acc)

    print(f"Epoch {epoch} => train_loss: {epoch_train_loss:.4f}, train_acc: {epoch_train_acc:.4f} | val_loss: {epoch_val_loss:.4f}, val_acc: {epoch_val_acc:.4f}")

    # Save best
    if epoch_val_acc > best_val_acc:
        best_val_acc = epoch_val_acc
        print(f"New best val_acc: {best_val_acc:.4f} — saving checkpoint to {BEST_PATH}")
        save_checkpoint(model, optimizer, epoch, BEST_PATH)

print("Training finished. Best val_acc:", best_val_acc)
```

```python
# Cell 1 - imports & helpers
import os, sys, numpy as np, torch, tqdm
import torch.nn as nn
import torch.nn.functional as F
from torchvision import datasets, transforms
import timm
import torchvision.models as tv_models
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# Helper: extract probable state_dict from saved checkpoint
def extract_state_dict(ckpt_path_or_obj):
    if isinstance(ckpt_path_or_obj, str):
        ck = torch.load(ckpt_path_or_obj, map_location='cpu')
    else:
        ck = ckpt_path_or_obj
    # Common keys
    for key in ("model_state", "model_state_dict", "state_dict", "model"):
        if isinstance(ck, dict) and key in ck:
            return ck[key]
    # If ck itself looks like a state dict
    if isinstance(ck, dict):
        # heuristic: keys are strings like "layer1.0.conv1.weight"
        if len(ck)>0 and all(isinstance(k, str) for k in list(ck.keys())[:10]):
            return ck
    raise ValueError("Could not extract state_dict from checkpoint: " + str(ckpt_path_or_obj))

# Generic builder + loader for timm/torchvision models (adapts final head for num_classes)
def build_and_load_model_entry(entry, num_classes):
    arch = entry.get("arch","timm")
    name = entry["model_name"]
    # build raw model object
    if arch == "timm":
        model = timm.create_model(name, pretrained=False, num_classes=num_classes)
    elif arch == "torchvision":
        mn = name.lower()
        if "resnet18" in mn:
            model = tv_models.resnet18(pretrained=False)
            model.fc = nn.Linear(model.fc.in_features, num_classes)
        elif "alexnet" in mn:
            model = tv_models.alexnet(pretrained=False)
            model.classifier[6] = nn.Linear(model.classifier[6].in_features, num_classes)
        elif "mobilenet_v2" in mn or "mobilenetv2" in mn:
            model = tv_models.mobilenet_v2(pretrained=False)
            model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)
        else:
            # fallback: try attribute in torchvision.models
            if hasattr(tv_models, name):
                model = getattr(tv_models, name)(pretrained=False)
                # best-effort replace final layer if present
                try:
                    if hasattr(model, "fc"):
                        model.fc = nn.Linear(model.fc.in_features, num_classes)
                    elif hasattr(model, "classifier") and isinstance(model.classifier, nn.Sequential):
                        last = list(model.classifier.children())[-1]
                        if isinstance(last, nn.Linear):
                            model.classifier = nn.Sequential(*list(model.classifier.children())[:-1], nn.Linear(last.in_features, num_classes))
                except Exception:
                    pass
            else:
                raise ValueError(f"Unknown torchvision model name: {name}")
    else:
        raise ValueError("Unknown arch: " + arch)

    ckpt = entry.get("ckpt")
    if ckpt:
        sd = extract_state_dict(ckpt)
        # tolerant loading (module. prefix handling)
        try:
            model.load_state_dict(sd)
        except RuntimeError:
            # try strip/add module.
            new_sd = {}
            for k,v in sd.items():
                if k.startswith("module."):
                    new_sd[k.replace("module.","")] = v
                else:
                    new_sd["module."+k] = v
            try:
                model.load_state_dict(new_sd)
            except Exception as e:
                # final attempt: if saved dict contains nested model_state
                raise RuntimeError(f"Failed to load state_dict for {entry.get('name','?')}: {e}")
    model.to(device)
    model.eval()
    return model

# Get input size expected by a model (try model.default_cfg if timm, else fallback 224)
def model_input_size(model):
    try:
        cfg = getattr(model, "default_cfg", None)
        if cfg and "input_size" in cfg and cfg["input_size"] is not None:
            s = int(cfg["input_size"][1])
            return s
    except Exception:
        pass
    # torchvision common default
    return 224

# Create an ImageFolder + restricted samples DataLoader with given transform for provided indices
from torch.utils.data import DataLoader, Subset

def make_loader_for_indices(root, indices, input_size, batch_size=32, num_workers=2):
    tf = transforms.Compose([
        transforms.Resize((input_size, input_size)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
    ])
    ds = datasets.ImageFolder(root=root, transform=tf)
    subset = Subset(ds, indices)
    loader = DataLoader(subset, batch_size=batch_size, shuffle=False, num_workers=num_workers, pin_memory=True)
    return loader

# Compute probability matrix (N x C) for a model and given loader
def compute_probs_for_model(model, loader):
    model.eval()
    all_probs = []
    all_labels = []
    with torch.no_grad():
        for xb, yb in loader:
            xb = xb.to(device)
            logits = model(xb)
            probs = F.softmax(logits, dim=1).cpu().numpy()
            all_probs.append(probs)
            all_labels.extend(yb.numpy().tolist())
    all_probs = np.vstack(all_probs)
    all_labels = np.array(all_labels)
    return all_probs, all_labels
```

```python
# Single cell: robustly find model checkpoints and load models (fixes FileNotFoundError)
import os, glob, re, torch, numpy as np
import torch.nn as nn
import timm
import torchvision.models as tv_models
from pathlib import Path
from collections import defaultdict

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ---------- CONFIG: adjust if your folders differ ----------
MODEL_ENTRIES = [
    {"name":"efficientnet_b7", "arch":"timm", "model_name":"tf_efficientnet_b7", "ckpt":"best_efficientnet_b7.pth"},
    {"name":"swinv2",         "arch":"timm", "model_name":"swinv2_large_window12_192_22k", "ckpt":"checkpoints/best_swinv2_large_window12_192_22k_acc0.9557.pth"},
    {"name":"resnet18",       "arch":"torchvision", "model_name":"resnet18", "ckpt":"/content/checkpoints_busi/resnet18_epoch_009.pth"},
    {"name":"alexnet",        "arch":"torchvision", "model_name":"alexnet", "ckpt":"/content/checkpoints_busi/alexnet_epoch_006.pth"},
    {"name":"mobilenet_v2",   "arch":"torchvision", "model_name":"mobilenet_v2", "ckpt":"/content/checkpoints_busi/mobilenetv2_epoch_008.pth"},
]

# directories to search for checkpoints (will search recursively)
SEARCH_DIRS = [
    ".", "./checkpoints", "./checkpoints_busi", "/content/checkpoints_busi", "/content/checkpoints",
    "/content/drive/MyDrive/BUSI_models", "/content/drive/MyDrive"
]

# helper to find candidate files for a model using heuristics
def find_ckpt_for_model(model_key):
    name = model_key.lower()
    patterns = []
    # high-priority patterns
    patterns += [f"**/*{name}*best*.pth", f"**/*{name}*best*.pt", f"**/*{name}*best*.ckpt"]
    patterns += [f"**/*{name}*acc*.pth", f"**/*{name}*acc*.pt"]
    # more general
    patterns += [f"**/*{name}*.pth", f"**/*{name}*.pt", f"**/*{name}*.ckpt", f"**/{name}*.pth", f"**/{name}*.pt"]
    # also check for generic 'final' or 'best' files that might not have model_name
    patterns += ["**/final*.pth", "**/final*.pt", "**/best*.pth", "**/best*.pt"]
    found = []
    for d in SEARCH_DIRS:
        d = os.path.expanduser(d)
        if not os.path.isdir(d):
            continue
        for pat in patterns:
            pathglob = os.path.join(d, pat)
            matches = glob.glob(pathglob, recursive=True)
            for m in matches:
                if os.path.isfile(m):
                    # skip files inside Drive target folder that may be copies
                    found.append(os.path.abspath(m))
    # deduplicate
    found = list(dict.fromkeys(found))
    return found

# scoring function to pick best candidate: prefer filename with acc value, then 'best' in name, then largest file
def score_ckpt_path(p):
    base = os.path.basename(p).lower()
    score = 0.0
    # prefer 'best' substring
    if "best" in base:
        score += 1000
    # prefer 'final'
    if "final" in base:
        score += 800
    # prefer acc pattern like acc0.9557 or acc09557
    m = re.search(r"acc[_\-]?([0-9]*\.?[0-9]+)", base)
    if m:
        try:
            score += float(m.group(1)) * 1e4
        except:
            pass
    # prefer epoch number if present
    m2 = re.search(r"epoch[_\-]?0*([0-9]{1,3})", base)
    if m2:
        try:
            score += int(m2.group(1))
        except:
            pass
    # largest file fallback: add size (will be small weight)
    try:
        score += os.path.getsize(p) / (1024*1024) * 0.1
    except:
        pass
    return score

# tolerant extractor for actual state_dict inside checkpoint
def extract_state_dict_from_file(p):
    ck = torch.load(p, map_location='cpu')
    if isinstance(ck, dict):
        for k in ("model_state", "model_state_dict", "state_dict", "model"):
            if k in ck:
                return ck[k]
        # maybe it's the state dict itself
        # heuristic: keys look like strings with dots
        keys = list(ck.keys())
        if keys and isinstance(keys[0], str) and ("." in keys[0] or "conv" in keys[0] or "layer" in keys[0] or "fc" in keys[0]):
            return ck
    # otherwise return ck (may be state_dict)
    return ck

# builder + loader (tolerant)
def build_model_and_load(entry, num_classes):
    arch = entry.get("arch", "timm")
    model_name = entry["model_name"]
    # create model instance
    if arch == "timm":
        model = timm.create_model(model_name, pretrained=False, num_classes=num_classes)
    else: # torchvision
        mn = model_name.lower()
        if "resnet18" in mn:
            m = tv_models.resnet18(pretrained=False)
            m.fc = nn.Linear(m.fc.in_features, num_classes)
            model = m
        elif "alexnet" in mn:
            m = tv_models.alexnet(pretrained=False)
            m.classifier[6] = nn.Linear(m.classifier[6].in_features, num_classes)
            model = m
        elif "mobilenet" in mn:
            m = tv_models.mobilenet_v2(pretrained=False)
            m.classifier[1] = nn.Linear(m.classifier[1].in_features, num_classes)
            model = m
        else:
            # try dynamic attr
            if hasattr(tv_models, model_name):
                model = getattr(tv_models, model_name)(pretrained=False)
                # try to adapt head if possible (best-effort)
                try:
                    if hasattr(model, "fc"):
                        model.fc = nn.Linear(model.fc.in_features, num_classes)
                    elif hasattr(model, "classifier") and isinstance(model.classifier, nn.Sequential):
                        last = list(model.classifier.children())[-1]
                        if isinstance(last, nn.Linear):
                            model.classifier = nn.Sequential(*list(model.classifier.children())[:-1], nn.Linear(last.in_features, num_classes))
                except:
                    pass
            else:
                raise ValueError("Unknown torchvision model: " + model_name)
    # load checkpoint if exists
    ckpt_path = entry.get("ckpt")
    if ckpt_path and os.path.isfile(ckpt_path):
        sd = extract_state_dict_from_file(ckpt_path)
        # try load
        try:
            model.load_state_dict(sd)
        except RuntimeError:
            # try strip/add module.
            new_sd = {}
            for k,v in sd.items():
                if k.startswith("module."):
                    new_sd[k.replace("module.","")] = v
                else:
                    new_sd["module."+k] = v
            model.load_state_dict(new_sd)
    model.to(device)
    model.eval()
    return model

# ----------------- main: find ckpts, pick best, load models -----------------
# Determine num_classes from dataset if available
try:
    NUM_CLASSES = len(dataset_all.classes)
except Exception:
    # fallback default (you can change)
    NUM_CLASSES = 3

print("Expecting NUM_CLASSES =", NUM_CLASSES)

resolved = {}
models = {}
skipped = []

for entry in MODEL_ENTRIES:
    name = entry["name"]
    requested_ckpt = entry.get("ckpt")
    print(f"\nResolving checkpoint for model '{name}' (requested: {requested_ckpt}) ...")

    # if requested path exists, accept it
    if requested_ckpt and os.path.isfile(requested_ckpt):
        chosen = os.path.abspath(requested_ckpt)
        print("  Requested path exists:", chosen)
    else:
        # search heuristically
        candidates = find_ckpt_for_model(entry.get("model_name") or name)
        # also add the originally requested path if it exists under other locations (e.g., maybe stored in DRIVE)
        if requested_ckpt:
            # try to glob the filename (basename) anywhere
            bn = os.path.basename(requested_ckpt)
            for d in SEARCH_DIRS:
                globpath = os.path.join(d, "**", bn)
                found = glob.glob(globpath, recursive=True)
                for f in found:
                    if os.path.isfile(f):
                        candidates.append(os.path.abspath(f))
        candidates = list(dict.fromkeys(candidates))
        if not candidates:
            print(f"  No checkpoint candidates found for '{name}'. Skipping load for this model.")
            skipped.append(name)
            continue
        # score and choose best
        scored = [(score_ckpt_path(p), p) for p in candidates]
        scored.sort(reverse=True)
        chosen = scored[0][1]
        print("  Found candidates:", len(candidates), "- choosing:", os.path.basename(chosen))

    # attempt to load/check state_dict quickly (and adjust if file holds nested dict)
    try:
        # build and load model (builder will also attempt to load if ckpt path present)
        entry_copy = dict(entry)
        entry_copy["ckpt"] = chosen
        m = build_model_and_load(entry_copy, NUM_CLASSES)
        models[name] = m
        resolved[name] = chosen
        print(f"  Loaded '{name}' successfully from -> {chosen}")
    except Exception as e:
        print(f"  FAILED to load model '{name}' from {chosen}: {e}")
        skipped.append(name)
        continue

print("\nSummary:")
print("  Loaded models:", list(models.keys()))
if resolved:
    print("  Resolved paths:")
    for k,v in resolved.items():
        print("   -", k, "->", v)
if skipped:
    print("  Skipped (no ckpt or load failed):", skipped)

# show input sizes (try model.default_cfg)
def model_input_size(m):
    try:
        cfg = getattr(m, "default_cfg", None)
        if cfg and "input_size" in cfg and cfg["input_size"]:
            return int(cfg["input_size"][1])
    except:
        pass
    return 224

for k,m in models.items():
    print(f"Model '{k}' input_size ->", model_input_size(m))
```

```python
# Cell 4 - compute probs on train (for CV) and test (final evaluation)
# We will tune weights on the TRAIN set using StratifiedKFold and evaluate the final ensemble on test indices.

# ensure train_idx/test_idx exist (from your earlier split)
try:
    train_idx, test_idx
except NameError:
    raise RuntimeError("train_idx / test_idx not found. Re-run the dataset-splitting cell to produce train_idx and test_idx.")

per_model_train_probs = {}
per_model_test_probs = {}
train_labels = None
test_labels = None

for name, m in models.items():
    print(">>> computing probs for model:", name)
    in_size = model_input_size(m)
    train_loader_for_model = make_loader_for_indices(DATA_ROOT, train_idx, input_size=in_size, batch_size=32)
    test_loader_for_model  = make_loader_for_indices(DATA_ROOT, test_idx,  input_size=in_size, batch_size=32)

    probs_train, labs_train = compute_probs_for_model(m, train_loader_for_model)
    probs_test,  labs_test  = compute_probs_for_model(m, test_loader_for_model)

    if train_labels is None:
        train_labels = labs_train
    else:
        assert np.array_equal(train_labels, labs_train), "train label mismatch!"

    if test_labels is None:
        test_labels = labs_test
    else:
        assert np.array_equal(test_labels, labs_test), "test label mismatch!"

    per_model_train_probs[name] = probs_train
    per_model_test_probs[name]  = probs_test

    print(f"  train probs shape: {probs_train.shape}, test probs shape: {probs_test.shape}")

print("Done. Train examples:", train_labels.shape[0], "Test examples:", test_labels.shape[0])
```

```python
# Cell 5 - grid-search weights (discrete grid) with StratifiedKFold CV on the TRAIN set
from itertools import product
import math

model_names = list(per_model_train_probs.keys())
M = len(model_names)
print("Ensembling models in this order:", model_names)

# grid resolution (e.g., 0.05 -> weights in increments of 0.05)
grid_step = 0.05
steps = int(round(1.0 / grid_step))

# generate combinations: integer partitions of `steps` across M buckets
combos = []
for tup in product(range(steps+1), repeat=M):
    if sum(tup) == steps:
        w = np.array(tup, dtype=float) / steps
        combos.append(w)
print("Total combos:", len(combos))

# prepare per-model train probs in a list (order matches model_names)
train_probs_list = [per_model_train_probs[n] for n in model_names]
y_train = train_labels

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

def cv_score(weights):
    # weights: numpy array length M, sum to 1
    accs = []
    for tr_idx, val_idx in cv.split(np.zeros(len(y_train)), y_train):
        # assemble ensemble probs on validation fold
        probs_val = np.zeros((len(val_idx), train_probs_list[0].shape[1]), dtype=float)
        for i,p in enumerate(train_probs_list):
            probs_val += weights[i] * p[val_idx]
        preds = probs_val.argmax(axis=1)
        accs.append(accuracy_score(y_train[val_idx], preds))
    return np.mean(accs)

# grid search
best_score = -1.0
best_weights = None
for w in combos:
    score = cv_score(w)
    if score > best_score:
        best_score = score
        best_weights = w

print("Best discrete-grid CV acc:", best_score)
for n,wt in zip(model_names, best_weights):
    print(f"  {n}: {wt:.3f}")
```

```python
# Cell 6 - optional refine using SciPy + evaluate on test set
try:
    from scipy.optimize import minimize

    def obj(x):
        ex = np.exp(x)
        w = ex / ex.sum()
        return -cv_score(w)

    x0 = np.log(np.maximum(best_weights, 1e-6))
    res = minimize(obj, x0, method='Powell', options={'maxiter':1000, 'disp': False})
    w_opt = np.exp(res.x) / np.exp(res.x).sum()
    opt_score = -res.fun
    print("Refined CV acc (SciPy):", opt_score)
    if opt_score > best_score + 1e-6:
        best_score = opt_score
        best_weights = w_opt
        print(" -> improvement found. Using refined weights.")
    else:
        print(" -> no improvement from refinement; keeping grid best.")
except Exception as e:
    print("SciPy refine skipped:", e)

print("\nFinal ensemble weights (sum {:.4f}), CV acc {:.4f}:".format(best_weights.sum(), best_score))
for n,w in zip(model_names, best_weights):
    print(f"  {n}: {w:.4f}")

# Evaluate on test set
test_probs_list = [per_model_test_probs[n] for n in model_names]
probs_ens_test = np.zeros_like(test_probs_list[0])
for i,p in enumerate(test_probs_list):
    probs_ens_test += best_weights[i] * p

y_pred_test = probs_ens_test.argmax(axis=1)
test_acc = accuracy_score(test_labels, y_pred_test)
print("\nEnsemble test accuracy:", test_acc)
print("\nClassification report on test set:")
print(classification_report(test_labels, y_pred_test, target_names=dataset_all.classes, digits=4))
print("\nConfusion matrix:\n", confusion_matrix(test_labels, y_pred_test))
```
